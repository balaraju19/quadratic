```python
import math
a = 1
b = -3
c = 2
dis = b**2 - 4*a*c
if dis >= 0:
    root1 = (-b + math.sqrt(dis)) / (2*a)
    root2 = (-b - math.sqrt(dis)) / (2*a)
    print(f"Roots: {root1}, {root2}")
else:
    print("no real roots")
```
